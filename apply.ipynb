```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df_books.head(2)
```

```python
#definicion de funcion
def two_times(value):
    return value * 2
```

```python
#Aplicar una funcion
df_books['User Rating'].apply(two_times)# a cada valor de la columna User Rating le aplica esta funci√≥n
```

```python
#Crear una nueva columna aplicando la funcon
df_books['Rating2']= df_books['User Rating'].apply(two_times)
df_books
```

```python
#aplicar funcion de tipo lambda
df_books['Rating2']= df_books['User Rating'].apply(lambda x: x *3)
df_books.head()
```

```python
#aplicar funcion a multiple columnas
#mult por 2 el User ratring 
df_books['Rating2'] = df_books.apply(lambda x : x['User Rating'] * 2 if x['Genre']=='Fiction' else x['User Rating'], axis=1)
#con axis = 1 se hace referencia alas columnas
df_books
```
